# Export simulated resonance signals for Gd

```python
import os
import sys
import re
import pprint
import numpy as np
```

```python
root_folder = os.path.dirname(os.getcwd())
sys.path.append(root_folder)
import ImagingReso
from ImagingReso.resonance import Resonance
```

```python
_energy_min = 0.0001
_energy_max = 300
_energy_step = 0.1
```

## Gd foil of 0.075 mm

```python
_layer_1 = 'Gd'
_thickness_1 = 0.075 # mm
```

```python
o_reso = Resonance(energy_min=_energy_min, energy_max=_energy_max, energy_step=_energy_step)
o_reso.add_layer(formula=_layer_1, thickness=_thickness_1)
```

```python
o_reso.plot(mixed=False, all_layers=True, y_axis='attenuation')
```

```python
o_reso.plot(mixed=False, all_layers=True, y_axis='transmission')
```

```python
o_reso.export(filename='Gd_75um.csv', mixed=False, all_layers=True, y_axis='transmission', to_csv=True)
```
